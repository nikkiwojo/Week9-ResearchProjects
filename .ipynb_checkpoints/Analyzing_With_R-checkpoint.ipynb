```r
# Import libraries
library('readr')
library('dplyr')
```

```r
# Read data from CSV file
data <- read.csv("employee_data.csv")
data
```

### Data Visualization

```r
# First, install the package ggplot2
install.packages('ggplot2')
```

```r
library('ggplot2')
```

```r
# Let's make a cool-looking dotplot that shows us the age of each employee

graph <- ggplot(data, aes(x = Age, y = Name)) # Makes the background of the graph
points <- geom_point(col='tomato2', size=3) # Draws the points in red
labels <- labs(title="Employee Ages", subtitle='Parks Dept') # Adds title to the graph

graph + points + labels # Puts it all together
```

```r
# Now let's make a bar graph to compare males vs. females employees

graph <- ggplot(data, aes(Gender))
bars <- geom_bar(width = 0.5, fill = 'darkorchid4')
lables <- labs(title = 'Genders of Employees', subtitle = 'Parks Dept')

graph + bars + lables
```

```r
# Let's do one more graph that compares two separate variables
# We'll have to import some new data

scores <- read.csv('student_scores.csv')
scores
```

```r
# Let's make a scatterplot to see if there is a correlation!

graph <- ggplot(scores, aes(x = Hours.Studied, y = Score))
points <- geom_jitter(col = 'green4')
labels <- labs(title='Scores vs. Study Time', subtitle = 'Is there a correlation?')

graph + points + labels
```

```r
# The following will add a linear regression to our graph

graph + points + labels + geom_smooth(method = 'lm')
```
